```python
data = sm.add_constant(data)
ivmod = IV2SLS(dependent = data['LogQuantity'], exog = data[['const','Monday','Tuesday','Wednesday','Thursday']], \
              endog = data['LogPrice'], instruments = data['Stormy'])
results_2sls = ivmod.fit(cov_type='unadjusted')
print(results_2sls)
```
